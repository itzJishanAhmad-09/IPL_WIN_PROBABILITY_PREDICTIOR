```python
import numpy as np
import pandas as pd

match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv/deliveries.csv')
```

```python
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
total_score_df = total_score_df[total_score_df['inning'] == 2]
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
match_df
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challenges Banglore',
    'Kolkata Knight Riders',
    'Kings XI punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df = match_df[['match_id','team1','team2','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.shape
```

```python
delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df['crr'] = delivery_df['current_score'] / (120 - (delivery_df['over'] * 6 + delivery_df['ball']))
```

```python
delivery_df['rrr'] = (delivery_df['runs_left'] * 6) / (120 - (delivery_df['over'] * 6 + delivery_df['ball']))
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
delivery_df['bowls_left'] = 120 - (delivery_df['over'] * 6 + delivery_df['ball'])
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','bowls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.dropna(inplace=True)
final_df = final_df[final_df['runs_left'] >= 0]
```

```python
from sklearn.model_selection import train_test_split
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'),['batting_team','bowling_team','city'])
]
, remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression())
])
```

```python

x_train = x_train.replace([np.inf, -np.inf], np.nan)

x_train = x_train.dropna()
y_train = y_train.loc[x_train.index]

pipe.fit(x_train, y_train)
```

```python
x_train.describe()
```

```python
x_test_clean = x_test.replace([np.inf, -np.inf], np.nan)
x_test_clean = x_test_clean.dropna()
y_pred = pipe.predict(x_test_clean)
```

```python
from sklearn.metrics import accuracy_score
y_test_clean = y_test.loc[x_test_clean.index]
accuracy_score(y_test_clean, y_pred)
```

```python
def match_progression(x_df, match_id, pipe):
    temp_df = x_df[x_df['match_id'] == match_id].copy()
    if temp_df.empty:
        return pd.DataFrame(), None
    temp_df = temp_df[temp_df['ball'] == 6].copy()
    if temp_df.empty:
        return pd.DataFrame(), None
    model_cols = ['batting_team', 'bowling_team', 'city', 'runs_left', 'bowls_left', 'wickets', 'total_runs_x', 'crr', 'rrr']
    model_df = temp_df[model_cols].copy()
    model_df = model_df.dropna()
    if model_df.empty:
        return pd.DataFrame(), None
    proba = pipe.predict_proba(model_df)
    temp_df = temp_df.loc[model_df.index]
    temp_df['loss'] = np.round(proba[:, 0] * 100, 1)
    temp_df['win'] = np.round(proba[:, 1] * 100, 1)
    temp_df['end_of_over'] = range(1, temp_df.shape[0] + 1)
    target = int(temp_df['total_runs_x'].iloc[0])
    temp_df['runs_after_over'] = target - temp_df['runs_left'].astype(int)
    wickets = temp_df['wickets'].astype(int).values
    prev_wickets = np.concatenate(([10], wickets[:-1]))
    wickets_in_over = prev_wickets - wickets
    temp_df['wickets_in_over'] = wickets_in_over
    result_df = temp_df[['end_of_over', 'runs_after_over', 'wickets_in_over', 'loss', 'win']].copy()
    return result_df, target
```

```python

temp_df, target = match_progression(delivery_df, 1, pipe)
```

```python
import matplotlib.pyplot as plt

if not temp_df.empty:
	plt.figure(figsize=(18,8))
	plt.plot(temp_df['end_of_over'], temp_df['wickets_in_over'], color='yellow', linewidth=3)
	plt.plot(temp_df['end_of_over'], temp_df['win'], color='#00a650', linewidth=4)
	plt.plot(temp_df['end_of_over'], temp_df['loss'], color='red', linewidth=4)
	plt.plot(temp_df['end_of_over'], temp_df['runs_after_over'])
	plt.title('target ' + str(target))
	plt.xlabel('End of Over')
	plt.legend(['Wickets in Over', 'Win %', 'Loss %', 'Runs After Over'])
	plt.show()
else:
	print("No data available for plotting. Please check the match_id or input data.")

```
